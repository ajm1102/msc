```python
import matplotlib.pyplot as plt
import os
import yfinance as yf
import json
import numpy as np

ticker = 'lly'

msft = yf.Ticker(ticker)

# get historical market data
hist = msft.history(period="1d", start="2023-02-15")
```

```python
data = []
pure_data = []
press_statements_classed = []

path_to_json = './data/'

for file_name in [file for file in os.listdir(path_to_json) if file.endswith('.json')]:
    with open(path_to_json + file_name, encoding="utf8") as json_file:
        file = json.load(json_file)
        data.extend(file)
```

```python
from datetime import datetime


for company in data:
    if company['ticker'] == ticker:
        company_data = company['data']
        
company_dates = [datetime.strptime(x['date'], '%Y-%m-%d').date() for x in company_data]
```

```python
import mplfinance as mpf


hist2 = hist.copy(deep=True)
hist2['date'] = hist.index.date
hist2[~hist2['date'].isin(company_dates)] = np.nan


apd = mpf.make_addplot(hist2['Low'], type='scatter')

config = dict(
    figratio=(20,12),
    tight_layout=False
)

plot = mpf.plot(hist,addplot=apd, **config)
```
